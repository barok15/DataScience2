```python
import pandas as pd 
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('ETdata.csv')
```

```python
sns.scatterplot(data=df, x='Schooling (average number of years completed)', y='Life Expectancy (Years)')
```

```python
df_train, df_test = train_test_split(df, test_size=0.3)
```

```python
def train(df_train):
    # y = mx+b
    # b = 0 
    
    X_train = df_train[['Schooling (average number of years completed)']]
    y_train = df_train['Life Expectancy (Years)']

    # Calculate the average change in life expectancy for a one-unit change in schooling
    average_slope = y_train.mean() / X_train.mean()

    return average_slope


```

```python
#US 2023 
# schooling = 13.5 years 
schooling_us_2023 = 13.3  # Assume 13.3 years of schooling

# Coefficients and intercept obtained from the trained model
slope = train(df_train)

print(slope)

life_US_2023 = (slope*schooling_us_2023)
print(life_US_2023)
```
