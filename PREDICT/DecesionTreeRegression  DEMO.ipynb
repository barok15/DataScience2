```python
import pandas as pd 
import seaborn as sns 
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("cars.csv")
```

```python
sns.scatterplot(data=df, x="city mpg", y="MSRP")
```

```python
makes = df['Make'].unique()
makes
```

```python
train, test = train_test_split(df, test_size = 0.3)
```

```python
def trainDT(df): 
    avgByMake = {}
    for make in makes: 
        avgByMake[make] = df[df['Make'] == make]["MSRP"].mean()
        
    return avgByMake

avgs = trainDT(df)
```

```python
# Root for Model 1: What make is it?
import math 
def testModel(df, model): 
    sum =  0 
    for idx in range(len(df)): 
        row = df.iloc[idx]
        make = row['Make']
        actual = row["MSRP"]
        predicted = model[make]
        error = (actual - predicted)**2
        sum = sum + error
        
    return math.sqrt(sum/len(df))
```

```python
testModel(test, avgs)
```
